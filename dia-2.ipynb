```python
# challenge 1
entrada_al_asuncionico = int(input('ingrese su edad'))

if  entrada_al_asuncionico <= 18:
    print('menor de edad, pase al area para chicos')

elif  entrada_al_asuncionico <= 21:
    print('acceso permitido, prohibida la compra de bebidas')

else:
    print('acceso completo, disfrute de nuestro asuncionico')
```

```python
 # challenge 2

temperatura_actual = int(input('digite la temperatura actual'))
    
if temperatura_actual >= -10 and temperatura_actual <= 13:
        print('ro`ykoko')

elif temperatura_actual >= 14 and temperatura_actual <= 23:
        print('refresca`i')  

elif temperatura_actual >= 24 and temperatura_actual <= 28:
        print('oikoite')

elif temperatura_actual >= 29 and temperatura_actual <= 45:
        print('hakuro')   

else:
        print('ñamano mba`e')  
```
